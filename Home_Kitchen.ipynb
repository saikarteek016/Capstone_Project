```python
sa = SentimentIntensityAnalyzer()
```

```python
def Sentiment(text):
  y = sa.polarity_scores(text)
  x = max(y,key=y.get)
  if x=='pos':
    return 'Positive'
  elif x=='neg':
    return 'Negative'
  else:
    return 'Neutral'
```

**Key drivers that effect people to purchase**

```python
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer()

import re
def clean_text(text):
    text = text.lower()
    re.sub(r"{","",text)
    re.sub(r"}","",text)
    re.sub(r"what's", "what is ", text)
    re.sub(r"'s", " ", text)
    re.sub(r"'ve", " have ", text)
    re.sub(r"can't", "can not ", text)
    re.sub(r"n't", " not ", text)
    re.sub(r"i'm", "i am ", text)
    re.sub(r"'re", " are ", text)
    re.sub(r"'d", " would ", text)
    re.sub(r"'ll", " will ", text)
    re.sub("[^0-9A-Za-z\-]+"," ",text) 
    re.sub("(?<!\w)\d+","",text)       
    re.sub("-(?<!\w)| (?<!\w)-",'',text) 
    
    text = text.strip(' ')

    return text
```

```python
from statsmodels.tsa.stattools import adfuller
def checkstationarity(data):
    pvalue=adfuller(data)[1]
    print(pvalue)
    if pvalue<0.05:
        msg='pvalue={}. Data is stationary, proceed to model building'.format(pvalue)
    else:
        msg='pvalue={} Data is not stationary, Make the data stationary before model building'.format(pvalue)
    
    return msg
```

1. What are the most common complaints or issues that customers have with the product?


```python
import nltk
nltk.download("stopwords")
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

# Tokenize the reviews and remove stop words
stop_words = set(stopwords.words('english'))
tokens = []
for review in hk_data['reviewText']:
    words = word_tokenize(review.lower())
    words = [word for word in words if word.isalnum() and word not in stop_words]
    tokens.extend(words)

complaints = ['quality', 'defective', 'broken', 'poor', 'not working', 'damaged']
complaint_count = Counter()
for token in tokens:
    if token in complaints:
        complaint_count[token] += 1

# Print the most common complaints/issues
print(complaint_count.most_common(5))
```

How frequently do customers mention the brand or manufacturer in the reviews?

```python
import pandas as pd
from collections import Counter

# Define a function to count the number of occurrences of a word in a text
def count_occurrences(text, word):
    return Counter(text.split())[word]

# Find the frequency of the brand or manufacturer in the reviews
brand_frequency = {}
for i, row in hk_data.iterrows():
    brand = row['brand']
    review_text = row['reviewText']
    if pd.notna(brand):
        if brand not in brand_frequency:
            brand_frequency[brand] = 0
        brand_frequency[brand] += count_occurrences(review_text, brand)

# Print the frequency of the brand or manufacturer in the reviews
if len(brand_frequency) > 0:
    print("Frequency of brand or manufacturer in the reviews:")
    for brand, frequency in brand_frequency.items():
        print(f"{brand}: {frequency}")
else:
    print("No brand or manufacturer mentioned in the reviews.")
```

What is the overall sentiment of the reviews? Are customers generally satisfied or dissatisfied with the product?

```python
import pandas as pd
from textblob import TextBlob

# Calculate the sentiment polarity for each review
sentiment_polarities = []
for i, row in hk_data.iterrows():
    review_text = row['reviewText']
    blob = TextBlob(review_text)
    sentiment_polarity = blob.sentiment.polarity
    sentiment_polarities.append(sentiment_polarity)

# Calculate the overall sentiment polarity for all reviews
overall_sentiment_polarity = sum(sentiment_polarities) / len(sentiment_polarities)

# Determine whether customers are generally satisfied or dissatisfied with the product
if overall_sentiment_polarity > 0:
    print("Customers are generally satisfied with the product.")
elif overall_sentiment_polarity < 0:
    print("Customers are generally dissatisfied with the product.")
else:
    print("Customers have mixed opinions about the product.")
```
